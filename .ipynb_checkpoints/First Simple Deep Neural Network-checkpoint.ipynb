In the introduction we have seen the components of the DDN.
We are going to put all this together and build our first simple Neural Network and then build from that.

```python
#we are going to use some data from the keras library: Boston housing
from keras.datasets import boston_housing
```

```python
(x_train, y_train), (x_test, y_test) = boston_housing.load_data()
```

```python
#let's see the way the data looks like
type(x_train)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
#what are the contents of the datasets
x_train[:5,]
```

```python
#since we have all the data as numerical now we continue to build our simple first model
from keras.models import Sequential #we are going to build a sequential model
from keras.layers import Dense, Activation #these will help input a dense later and an activation function

#Extract the last 100 rows from the training data to create the validation datasets.
x_val = x_train[300:,]
y_val = y_train[300:,]
```

```python
#Define the model architecture
model = Sequential() #- create sequential model
model.add(Dense(13, input_dim=13, kernel_initializer='normal',activation='relu'))#has 13 inputs and normal weight generalization
model.add(Dense(7, kernel_initializer='normal',activation='relu'))#create dense layer with 6 neurons
model.add(Dense(1, kernel_initializer='normal'))#create dense later with 4 neurons
```

```python
# Compile model
model.compile(loss='mean_squared_error', optimizer='adam',metrics=['mean_absolute_percentage_error'])
```

```python
#Train the model
model.fit(x_train, y_train, batch_size=64, epochs=1000,validation_data=(x_val,y_val))
```

```python
results = model.evaluate(x_test, y_test)
```

```python
print('Test accuracy: ', results[1])
```

We can conclude that the metric we are using keeps reducing with time. After 1000 epochs the error has significantly reduced to 0.16 which is great.
